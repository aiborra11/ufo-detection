```python
import pandas as pd
import numpy as np
import folium
```

```python
data = pd.read_csv('~/Desktop/Git-iron/ufo-lab/ufo.csv', engine='python')
print(data.columns)
display(data.head())
```

```python
data.dtypes
```

```python
def data_cleaner(df):
    data_old = df[df['year'] >= 2004]
    data_new = data_old[data_old['country'] == 'us']
    return data_new

data_older = data_cleaner(data).dropna()
```

```python
def data_cleaner2(df):
    data_young = df[df['year'] < 2004]
    data_new = data_young[data_young['country'] == 'us']
    return data_new

data_younger = data_cleaner2(data).dropna()
```

```python
from folium.plugins import HeatMap
```

```python
data_HM_younger = data_younger[['latitude','longitude' ]]
```

```python
m = folium.Map([40, -90], tiles='stamentoner', zoom_start=4)

HeatMap(data_HM_younger).add_to(folium.FeatureGroup(name='Heat Map').add_to(m))
folium.LayerControl().add_to(m)
m
```
